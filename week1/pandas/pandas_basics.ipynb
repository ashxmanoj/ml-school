There are two basic object types in pandas that hold data
- Data frame is just a table with rows & columns
- Each of those column is known as a pandas series
- Pandas series are iterable just like lists, strings

```python
# Select a pandas series from a df
df["Name"]  # Bracket notation
print(type(df['Name']))

df.Name  # Dot notation
```

But there are few limitations to dot notation
- If series name has a space in it
- We cannot use this method if series name clashes with a built-in attribute or method

```python
# Rename the column names
df.rename(columns={'Sp. Def' : 'Sp_defense', 'Sp. Atk' : 'Sp_attack'}, inplace=True)
```

```python
# Remove columns from a pandas dataframe
df.drop('Type 2', axis=1)
```

```python
df.drop(["Type 1", 'Type 2'], axis=1)
```

```python
# Drop rows using drop() method
df.drop(2, axis=0).head()
```

```python
# Sorting a pandas series
df.Name.sort_values(ascending=True)  # doesnot change the original 'Name' series
```

```python
# Sorting the df by a series
df.sort_values('Attack', ascending=False)
```

```python
# Filter rows of a pandas dataframe by a column value
booleans = []
for i in df.Attack:
    if i >= 180:
        booleans.append(True)
    else:
        booleans.append(False)

booleans_atk = pd.Series(booleans)
```

```python
df[df.Attack >= 180]
```

```python
# Multiple filter criteria
# df[(df.Attack >= 170) | (df.Defense >= 100)]
df[(df.Attack >= 170) & (df.Defense >= 100)]
```

- axis = 0 > Move across the row axis
